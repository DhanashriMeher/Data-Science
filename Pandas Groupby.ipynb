# Aggregate Operations

```python
df.dtypes
```

```python
nat
```

```python
x=nat[['CO2 Emissions(g/km)','Fuel Consumption Comb (mpg)']]
x
```

```python
x.agg(sum)
```

```python
x.agg([sum,min,max]) #Add multiple function. 
```

```python
#using numpy in agg
import numpy as npp
x.agg([npp.mean,npp.std])
```

```python
df.groupby('Make').agg([npp.mean,npp.max,npp.min])[['Engine Size(L)','CO2 Emissions(g/km)']]
```

```python
#convert['Z'=0,'D'=1,'X'=2,'E'=3,'N'=4]
df['Fuel Type'].map({'Z':0,'D':1,'X':2,'E':3,'N':4})
```

```python
#add new column for fuel Type with number replacement
df['new_ft']=df['Fuel Type'].map({'Z':0,'D':1,'X':2,'E':3,'N':4})
df['new_ft']
```

```python
#change original column by new vvalues
df['Fuel Type']=df['Fuel Type'].map({'Z':0,'D':1,'X':2,'E':3,'N':4})
df['Fuel Type']                   
```
