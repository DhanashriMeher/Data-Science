# Histogram

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv('PimaIndiansDiabetes.csv')
df
```

```python
plt.hist(df['BloodPrs'])
```

```python
#Specify own range
plt.hist(df['BloodPrs'],bins=[30,40,50,60,70,80,100,120],color ='red',alpha=0.4)
```

```python
#skip extra information about plot
#Specify own range
plt.hist(df['BloodPrs'],bins=[30,40,50,60,70,80,100,120],color ='red',alpha=0.4,label='Bp Range')
plt.legend()
plt.show()
```

```python
#to get histogram on all numeric columns
df.hist(figsize=(12,10))
plt.show()
```

### BarPlot

-Bivatiate
-Category(X axis)vs count(Y axis)display

```python
data1.index #x axis
```

```python
data1.values #y axis
```

```python
plt.bar([0,1],data1.values,width=0.5,color=['red','blue'])
plt.xticks([0,1],['Non-Diabetic','Diabetic'],rotation=90,);
#through ; we can hide rest information.
```

```python
#add two bar plots in single graph

x=['Group1','Group2','Group3','Group4','Group5']
y=df['BMI'].head(5)
z=df['Age'].head(5)

x_axis =np.arange(len(x))

plt.bar(x_axis - 0.2,y,width = 0.4,label ='BMI Count')
plt.bar(x_axis + 0.2,z,width = 0.4,label ='Age Count')
plt.legend()
plt.show
```
