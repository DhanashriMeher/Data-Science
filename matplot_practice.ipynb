## Univariate Plot

```python
plt.plot(df['Wage'])
```

```python
data=df.groupby('Skill Moves')['Potential'].mean()
data
```

```python
plt.plot(data,marker='*',color='red',linestyle='--',label='Potential')
#plt.text(4,100,'FIFA_GRAPH')
plt.xlabel('Skills')
plt.ylabel('Potential')
plt.grid()
plt.legend()
```

```python
#Combined two plots
Weight = df['Weight']
Potential =df['Potential']
plt.plot(Weight,label ='plot1')
plt.plot(Potential,label ='plot2')
plt.legend()
```
