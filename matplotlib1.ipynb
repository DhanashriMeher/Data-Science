# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  #pyplot is submodule for 2D plotting
```

```python
df=pd.read_csv('Canada.csv')
df.head()
```

#Line Plot
- Univariate,Bivariate,Multivariate
- When we need to trace pattern,treds over the period.
Ex. Stock Market

```python
df.columns
```

```python
#Univariate Plot
plt.plot(df['Fuel Consumption Hwy (L/100 km)'])
```

```python
plt.plot(df['CO2 Emissions(g/km)'])
```

```python
plt.plot([20,60,10,150])
```

```python
#group cylinders
data=df.groupby('Cylinders')['CO2 Emissions(g/km)'].mean()
data
```

```python
plt.plot(data,marker='*',color='red',linestyle='--',label='CO2 Emission')
plt.text(4,200,'Dhanashri')
plt.title('CO2 Emission per cylinder')
plt.xlabel('Cylinders')
plt.ylabel('CO2 Emission')
plt.grid()
plt.legend()

#to save the plot
plt.savefig('p.jpg')
```

```python
#Combine two plots
plt.plot([10,22,30,40],label ='plot1')
plt.plot([2,40,16,33],label='plot2')
plt.legend()
```

```python
#Separate 2 plot:one after another
plt.plot([10,22,3,45],label='plot1')
plt.legend()
plt.show()

plt.plot([2,40,16,33],label='plot2')
plt.ylim((1,50))
plt.legend()
plt.show()
```
